```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
 file = 'soma.wav.ogg'
```

```python
import IPython.display as ipd
```

```python
ipd.Audio(file)
```

```python
y, sr = librosa.load(file)
librosa.display.waveshow(y)
```

```python
from IPython.display import Audio
start_time = 1
end_time = 2
segment = y[int(start_time * sr):int(end_time * sr)]
Audio(segment, rate=sr)
```

```python
librosa.display.waveshow(segment, sr=sr)
```

```python
segment1 = y[:int(1 * sr)]
Audio(segment1, rate=sr)
```

```python
librosa.display.waveshow(segment1, sr=sr)
```

```python
segment2 = y[:int(1.5 * sr)]
Audio(segment1, rate=sr)
```

```python
librosa.display.waveshow(segment2, sr=sr)
```

```python
segment3 = y[:int(2 * sr)]
Audio(segment2, rate=sr)
```
