//A1. Find the first derivative of your speech signal with finite difference method. Listen to the first
derivative signal and the original speech signal.

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import librosa.display
```

```python
 file = 'soma.wav.ogg'
```

```python
import IPython.display as ipd
```

```python
ipd.Audio(file)
```

```python
y, sr = librosa.load(file)
librosa.display.waveshow(y)
plt.title('Speech Signal Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
from IPython.display import Audio
first_derivative = np.diff(y)
Audio(first_derivative, rate=sr)
```

```python
time_derivative = np.linspace(0, len(first_derivative)/sr, len(first_derivative))
plt.figure(figsize=(10, 4))
plt.plot(time_derivative, first_derivative)
```

A2. Detect the points of zero crossing in the first derivative signal. Compare the average length
between two consecutive zero crossings for speech and silence regions. Observe the pattern.


```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import librosa.display
```

```python
 file = 'soma.wav.ogg'
```

```python
import IPython.display as ipd
```

```python
ipd.Audio(file)
```

```python
zero_crossings_indices = np.where(np.diff(np.sign(first_derivative)))[0]
zero_crossings_diff = np.diff(zero_crossings_indices)
average_length = np.mean(zero_crossings_diff) / sr
print("zero_crossings_indices:",zero_crossings_indices)
print("zero_crossings_diff:",zero_crossings_diff)
print("Average length between two consecutive zero crossings (seconds):", average_length)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(time_derivative, first_derivative, label='First Derivative')
plt.scatter(time_derivative[zero_crossings_indices], first_derivative[zero_crossings_indices], color='black', label='Zero Crossings')
```

A4. Select a sentence which can be used for making a statement or asking a question. Ex: “You are
going to college on Sunday(./?)”. Record two signals – one with making the statement while other
with asking question. Study the two signals and compare them.

```python

statement_filename = 'answer.wav'  
question_filename = 'question.wav'
```

```python
statement_y, statement_sr = librosa.load(statement_filename)
question_y, question_sr = librosa.load(question_filename)

plt.figure(figsize=(10, 4))
plt.plot(statement_y)
plt.title('Statement Signal Waveform')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()


plt.figure(figsize=(10, 4))
plt.plot(question_y)
plt.title("Question Signal Waveform")
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()

statement_duration = librosa.get_duration(y=statement_y, sr=statement_sr)
question_duration = librosa.get_duration(y=question_y, sr=question_sr)

print("Duration of statement signal:", statement_duration, "seconds")
print("Duration of question signal:", question_duration, "seconds")
```
